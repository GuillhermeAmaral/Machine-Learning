# Atenuação de Ruído em Imagens Utilizando Stacked Autoencoders

Neste trabalho faremos uma rede neural autoencoder para atenuar o ruído de imagens em escala de cinza no banco de dados MNIST. Para implementar a rede usamos o Keras, que também nos fornece uma função para baixar as imagens necessárias.


## Autoencoders
Autoencoder é um algoritmo de compressão de dados, assim como JPEG e MP3 por exemplo, mas utilizando redes neurais de múltiplas camadas. Assim como ocorre nos exemplos citados, a função que realiza o processamento dos dados produz um resultado de menor qualidade em relação ao original. O que diferencia os autoencoders do resto são as seguintes características:
1. Autoencoders funcionam especificamente para os dados no quais foram usados para treinar a rede.. Logo, os dados de treino devem condizer com sua aplicação. Treinar a rede com imagens de faces lhe ensinará a comprimir apenas este objeto.
2. Autoencoders aprendem sozinhas a função de compressão a partir dos exemplos fornecidos e corretamente classificados.

O item 2 apresenta uma propriedade interessante: ela implica que é possível realizar compressão de mídias mesmo sem conhecimentos na área de processamento de imagens. O essencial é ter o conjunto de treino adequado e saber construir uma rede neural. 

Abaixo, um esquema de um autoencoder:
![SAE](img/SAE.jpeg)
Ao trabalhar-se com uma imagem, a entrada da rede são os valores de seus pixels. Na sessão do **codificador**, as camadas vão ficando menores a medida que os valores avançam. A idéia é que a informação seja representada por um número menor de valores. Na sessão do **decodificador**, ocorre o processo inverso e os poucos valores são usados para reconstruir a quantidade original de pixels. 

Como pode-se notar, a rede é treinada simplesmente com um conjunto de dados no qual a resposta desejada é igual à entrada. Notamos também que a saída de um autoencoder nunca é perfeitamente idêntica a entrada, mas deve ser o mais próximo possível.

Atualmente autoencoders não são muito usados em aplicações práticas, parte pela dificuldade de treinamento e parte por não serem aplicáveis a muitos tipos de imagens. Essas limitações não são problemas para os outro algoritmos existentes. Na verdade, eles têm sido usados mais para remoção de ruídos e redução de dimensionalidade para visualização de dados.

## Bibliotecas

```python
from keras.datasets import mnist
from keras.models import Model
from keras.layers import Input, Dense
import numpy as np
import matplotlib.pyplot as plt
```

A obtenção das imagens que servirão como dados de treino e teste é feita da seguinte forma:

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
plt.figure(figsize=(15,5))
for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title('Número {}'.format(y_train[i]))

plt.show()
```

Cada imagem é representada por uma matriz 28x28 com valores no intervalo [0, 255], como mostra a figura abaixo.

Vamos normalizar os dados para que eles possam ser trabalhados, após a normalização os valores dos pixels estarão no intervalo [0, 1].

```python
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
```

As camadas da rede neural pode ser pensada como um array, portanto a matriz que representa a imagem de um número será formatada para um array de tamanho igual a $28x28 = 784$

```python
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))
```

O problema proposto é o de conseguir recuperar uma imagem que contém ruído gaussiano, mas as imagens do dataset não possuem qualquer ruído, sendo assim ele será adicionado manualmente, como mostra o código abaixo:

```python
noise_factor = 0.5
x_train_noisy = x_train + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape) 
x_test_noisy = x_test + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape) 
x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)
```

Abaixo é mostrado o plot dos vinte imagens divididas em duas linhas. Na primeira tem-se as imagens dos dados de teste originais sem qualquer alteração e na segunda linha os dados com ruído.

```python
n = 10  # Quantidade de imagens que se deseja mostrar
plt.figure(figsize=(20, 4))
for i in range(n):
    # display original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display noisy
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(x_test_noisy[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```

A arquitetura da rede é a seguinte: 784 -> 128 -> 64 -> 32 -> 64 -> 128 -> 784. A implementação em código é mostrada abaixo.

```python
input_img = Input(shape=(784,))
encoded = Dense(128, activation='relu')(input_img)
encoded = Dense(64, activation='relu')(encoded)
encoded = Dense(32, activation='relu')(encoded)

decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(128, activation='relu')(decoded)
decoded = Dense(784, activation='sigmoid')(decoded)
```

O modelo e o treinamento é feito como mostrado abaixo. O otimizador ajusta a taxa de treinamento. Nessa arquitetura de rede neural, a entrada é a imagem com ruído e a saída é a correspondente sem ruído, dessa forma espera-se que a rede neural aprenda a como remover o ruído da imagem.

```python
autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
H = autoencoder.fit(x_train_noisy, x_train,
                epochs=100,
                batch_size=256,
                shuffle=True,
                validation_data=(x_test_noisy, x_test))
```

```python
plt.plot(H.history['loss'])
plt.title('Histórico de custo')
plt.xlabel('Época')
plt.xlabel('Custo')
plt.show()
```

O trecho de código abaixo mostra como salvar e ler os pesos obtidos após o treinamento.

```python
# SALVAR
W = autoencoder.get_weights()
np.save('SAE_Weight.npy', W)
# LER e configurar os pesos da rede obtidos em um treinamento anterior
#W = np.load('SAE_Weight.npy')
#autoencoder.set_weights(W)
```

O resultado é obtido após a codificação e decodificação da imagem, esse processo é mostrado abaixo.

```python
encoder = Model(input_img, encoded)
encoded_imgs = encoder.predict(x_test)

encoded_input = Input(shape=(32,))
decoder = Model(encoded_input, autoencoder.layers[-3](encoded_input))
decoded_imgs = decoder.predict(encoded_imgs)

encoded_input = Input(shape=(64,))
decoder = Model(encoded_input, autoencoder.layers[-2](encoded_input))
decoded_imgs = decoder.predict(decoded_imgs)

encoded_input = Input(shape=(128,))
decoder = Model(encoded_input, autoencoder.layers[-1](encoded_input))
decoded_imgs = decoder.predict(decoded_imgs)
```

A figura abaixo mostra o plot onde na primeira linha tem-se a imagem original, na segunda linha a imagem com ruído e na terceira e última linha a imagem após ser processada pela rede neural.

```python
n = 10  # how many digits we will display
plt.figure(figsize=(20, 4))
for i in range(n):
    # display original
    ax = plt.subplot(3, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display noise
    ax = plt.subplot(3, n, i + 1 + n)
    plt.imshow(x_test_noisy[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display reconstruction
    ax = plt.subplot(3, n, i + 1 + 2*n)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```
